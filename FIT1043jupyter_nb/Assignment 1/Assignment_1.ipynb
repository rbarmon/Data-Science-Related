#### 2. Create a line plot showing the total number of annual smartcard replacements (number of transactions) against year

To create a line plot showing the total number of annual smartcard replacements (number of transactions) against year, we first group the dataframe by Year and then sum the number of transactions. We then plot this new dataframe as a line plot shown below.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#total number of annual smartcard replacements (number of transactions) vs. year
df = smartcard.groupby('Year')['Number.of.transactions'].sum()
```

```python
df 
```

```python
plt.plot(df)
plt.xlabel('Year')
plt.title('Year vs. Annual Smartcard Replacements')
plt.ylabel('Annual Smartcard Replacements')
plt.show()
```

To find the annual number of transactions for each 'Action.Reason' across different years we need to groupby Action Reason and the Year column we generated and again sum the Number of Transactions. Which we achieve in the below code:

#### 2. For each action reason determine the number of years during which the number of annual transactions exceeds 10000.

To determine the number of years during which the number of annual transactions exceeds 10000 for each action reason, we take the data frame generated from above and take the rows that exceed 10000 based on a conditional. This will return rows with 0 or 1 based on whether the annual transaction exceeds 10000 so then we group by Action Reason and sum the 0s and 1s which will return the following data frame that shows for each action reason the number of years during which the number of annual transactions exceeds 10000.

We can observe that: <br>

```python
df2 = df.unstack()
df3 = (df2 > 10000).groupby('Action.Reason').sum()
df3
```

#### 4. Create a histogram to analyse the distribution of the annual number of transactions per action reason as calculated in A5.1. Explain any observations and provide comments on the distribution.

To create a histogram to analyse the distribution of the annual number of transactions per action reason as calculated in A5.1, we get the dataframe from A5.1 and reset the index to pivot the table into having the year as the index and the action reason as columns and the values as the number of transactions. This makes it possble to make the histogram based on the action reason which we plot as below.

```python
df = smartcard.groupby(['Action.Reason', 'Year'])['Number.of.transactions'].sum().reset_index()
df3 = df.pivot(index= "Year", columns = "Action.Reason", values="Number.of.transactions")
df3
```

To comment on the distribution it can be seen that the Action Reasons such as Condition Change and Lost tend to be consistent in being a reason for replacement and other minor reasons such as Disaster Relief and Remove Gender from Smartcard tend to be consistently low. So one observation that can be said is that common replacement reasons tend to remain common replacement reasons and vice versa. The overall ratio of each action reason also seems to be pretty consistent across the years which aligns with the first observation and this detail can also be seen the bar chart below. 

```python
df3.plot(kind='hist', bins=100, xlabel = 'Annual Number of Transactions', ylabel='Frequency', title='Frequency of Annual number of transactions per Action reason')
```
