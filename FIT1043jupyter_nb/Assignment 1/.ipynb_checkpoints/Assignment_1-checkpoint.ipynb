### A1. Exploring Smartcard Types

#### 1. How many different (unique) smartcard types are recorded in the ‘Smartcard.Type’ column? What are those different smartcard types and how many instances are recorded for each type?

```python
import pandas as pd 
smartcard = pd.read_csv("monthly_smartcard_replacements.csv")
```

Find the number of unique smartcard types that are recorded in the ‘Smartcard.Type’ column by using the nunique() function. 
<br><br>
<b>
From the code below, there are four unique smartcard types recorded in the ‘Smartcard.Type’ column.

```python
smartcard['Smartcard.Type'].nunique()
```

Now to derive the names of the different smartcard types the function unique() is used. 

<br><b>The code below prints that the names of the different smartcard types are 'Photo Identification Card', 'Driver Licence Card', 'Industry Authority Card', and 'Marine Licence Ind Card'.

```python
print(smartcard['Smartcard.Type'].unique())
```

To derive how many instances are recorded for each type we use the value_counts() function to count the number of instances for each type. <br>
<br>
<b>
From the below code it be observed that:<br>
    the Driver Licence Card has 1896 instances,<br>
    the Photo Identification Card has 1631 instances,<br>
    the Industry Authority Card has 1218 instances, and <br>
    the Marine Licence Ind Card has 1047 instances.

```python
print(smartcard['Smartcard.Type'].value_counts())
```

#### 2. Plot a barchart of the smartcard types with the bars showing the count of each smartcard type. Which smartcard type is replaced the most?

To plot a barchart of the smartcard types with the bars showing the count of each smartcard type we use matplotlib.pyplot and get the value_counts() which we generated above and then plot the details with a bar graph.<br>

<b>From the barchart below it can be observed that the driver license card is replaced the most because it has the most instances of being replaced. 

```python
import matplotlib.pyplot as plt
%matplotlib inline
smartcard['Smartcard.Type'].value_counts().plot(kind='bar', title='Replacement Count of Smartcard Types', xlabel='Smartcard Type', ylabel='Count')
```

#### 3. Calculate the percentage of records for each smartcard type.

To generate the percentage of records for each smartcard type we first used the value_counts() function as before. Then we add the normalize=True parameter which will return the relative frequencies of generated value count which is basically the percentage in decimal form so we then multiply the relative frequecies by 100 and round it to the first decimal place. Finally we convert it to a string to concatenate the percentage sign and then we have the results generated from the code below.<br>

<b> From the code below it can be observed that there are:<br>
32.7% of Driver Licence Card records,<br>
28.2% of Photo Identification Card records,<br>
21.0% of Industry Authority Card records, and<br>
18.1% Marine Licence Ind Card records.<b>

```python
smartcard['Smartcard.Type'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
```

### A3. Investigating Annual Smartcard Replacements

#### 1. Create a new column named 'Year' extracting the year from the 'Month' column.
Hint: you can extract year from column ‘Month’ using method .dt.year and create a
new column for year as follows:
>>> your_dataframe['Year']=your_dataframe['Month'].dt.year

To create a new column named 'Year' by extracting the year from the 'Month' column, we first extract the year from the 'Month' column by using the following code smartcard['Month'].dt.year, which takes the 'Month' column and creates a series of years by extracting it from the datetime in the 'Month' column. Then this series is assigned as a new column called 'Year'. The resulting new column is shown below.

```python
smartcard['Year']=smartcard['Month'].dt.year
```

```python
smartcard
```

#### 2. Create a line plot showing the total number of annual smartcard replacements (number of transactions) against year

To create a line plot showing the total number of annual smartcard replacements (number of transactions) against year, we first group the dataframe by Year and then sum the number of transactions. We then plot this new dataframe as a line plot shown below.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#total number of annual smartcard replacements (number of transactions) vs. year
df = smartcard.groupby('Year')['Number.of.transactions'].sum()
```

```python
df 
```

```python
plt.plot(df)
plt.xlabel('Year')
plt.title('Year vs. Annual Smartcard Replacements')
plt.ylabel('Annual Smartcard Replacements')
plt.show()
```

#### 3. Explain the trend as observed from the chart. Are there any years that are different from others and if so, what is the reason behind it?

The trend observed from the chart shows that initial number of replacements were low in 2010 when it was perhaps still at the early stages since the release of the Smartcards. From then to 2021 as more people began using smart cards the number of replacements increased drastically perhaps from a boom.
